# Imputer

_En Data Science, il arrive souvent qu’il y ait des valeurs manquantes dans un dataset. Voyons comment on peut créer une classe qui va nous permettre de remplacer cette valeur manquante par la moyenne des valeurs de la liste_

1. _Créez une classe qu’on appellera Imputer_
2. _Pour simplifier l’exercice, on ne traitera qu’avec des listes pour l’instant_
3. _Notre classe prendra un attribut qu’on appellera liste_
4. _Créez une fonction avg() qui d’abord retirera la valeur manquante puis la remplacera par la moyenne de la liste_

```python
class Imputer:
  def __init__(self, lst):
    self.liste = lst
  
  def avg(self):
    print("Voici ma liste intiale {}".format(self.liste))
    noneCount = self.liste.count(None)
    somme = 0
    for item in self.liste:
      if isinstance(item, (int, float)):
        somme += item
    average = somme/(len(self.liste)-noneCount)
    tempList = [average if item == None else item for item in self.liste] 
    print("Voici la liste avec les valeurs vide remplacées par la moyenne {}".format(tempList))
  
  def median(self):
    print("Voici ma liste intiale {}".format(self.liste))
    tempList = [item for item in self.liste if item != None]
    tempList.sort()
    median = 0
    countList = len(tempList)
    if countList % 2 == 0 :
      countList = int(countList/2)
      median = (tempList[countList-1]+tempList[countList])/2
    else:
      countList = int(((countList+1)/2)-1)
      median = tempList[countList]
    tempList = [median if item == None else item for item in self.liste] 
    print("Voici la liste avec les valeurs vide remplacées par la médiane {}".format(tempList))

imputer = Imputer([None, 2, 4, 6, None])
imputer.avg()
imputer2 = Imputer([10, 2, 4, 6, None])
imputer2.avg()
imputer3 = Imputer([5.6, 2, 3, 12, 5, 6, 5.6])
imputer3.avg()

```
