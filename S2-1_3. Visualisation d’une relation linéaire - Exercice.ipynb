2. On voudrait connaître la probabilité de partir de la l’entreprise par rapport à la distance par rapport à la maison. Utilisez _lmplot()_ pour visualiser cela. Que pouvez vous conclure ?

```
dataset = pd.read_csv('/content/drive/My Drive/IBM_HR_ATTRITION.csv')
dataset.Attrition = dataset.Attrition.apply(lambda x: 1 if x.lower()=="yes" else 0)
sns.lmplot(x="DistanceFromHome", y="Attrition", data=dataset)
```
