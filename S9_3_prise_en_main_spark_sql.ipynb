```
applestore = applestore.withColumn("%rating_count_ver", F.col("rating_count_ver")/F.col("rating_count_tot"))
```

```
display(applestore.groupBy("prime_genre").agg(
F.count("track_name"),
F.countDistinct("cont_rating"),
F.sum("rating_count_tot"),
F.mean("price"),
F.stddev("price")
))
```

```
applestore_cat = applestore.groupBy("prime_genre")
```

```
display(applestore.groupBy("prime_genre").agg(F.collect_list("track_name")))
```

```
display(applestore.select(
"track_name",
F.split(F.col("track_name"), " ").alias("words_list"),
F.explode(F.split(F.col("track_name"), " ")).alias("exploded_list")
)
       )
```

```
display(
  applestore.select(
    "prime_genre",
    F.explode(F.split(F.col("track_name")," ")).alias("word")
  ).groupBy("prime_genre")
  .agg(F.collect_set("word").alias("words"))
)
```

```
applestore_rank = applestore.withColumn("rank",
                                        F.rank().over(
                                        Window.partitionBy("prime_genre")
                                        .orderBy(F.desc("size_bytes"))
                                        )
                                       )
display(applestore_rank.select("track_name", "prime_genre", "rank", "size_bytes").orderBy("rank"))
```
