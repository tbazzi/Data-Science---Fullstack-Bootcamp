#2. Sélection et filtrage

**Equivalent SQL:**

**SELECT** track_name, currency, price
**WHERE** prime_genre = 'Games'
**FROM** apple_store

**Equivalent Pandas:**

apple_store.loc[ apple_store['prime_genre'] == 'Games', ['track_name', 'currency', 'price'] ]

```
import pyspark.sql.functions as F

display(
  apple_store.select("track_name", "currency", "price", "prime_genre")
             .filter(F.col("prime_genre") == "Games")
)
```

**Equivalent SQL:**

**SELECT** track_name, currency, price, prime_genre
**WHERE** prime_genre **NOT IN** ('Games', 'Weather')
**FROM** apple_store

**Equivalent Pandas:**

apple_store.loc[ ~apple_store['prime_genre'].isin(['Games', 'Weather']), ['track_name', 'currency', 'price', 'prime_genre'] ]

```
display(
  apple_store.select("track_name", "currency", "price", "prime_genre")
             .filter(~F.col("prime_genre").isin(["Games", "Weather"]))
)
```

#3. Transformations sur les colonnes

##A. Opérations simples

###Ajouter une colonne

```
apple_store_with_eu_prices = apple_store.withColumn("price_euros", F.col('price')*0.86)
```

On peut également ajouter une colonne avec la fonction select

```
apple_store_with_eu_prices = apple_store.select(
  *[F.col(col) for col in apple_store.columns],
  (F.col('price')*0.86).alias("price_euros")
)
```

###Ajouter une constante dans une colonne

```
# Ne fonctionne pas : 
# apple_store = apple_store.withColumn("Store", "Apple")

apple_store = apple_store.withColumn("Store", F.lit("Apple"))
```

###Effectuer une opération sur plusieurs colonnes

```
apple_store = apple_store.withColumn(
    "% rating_count_ver", 
    100 * F.col("rating_count_ver") / F.col("rating_count_tot")
  )
```

##B. Fonctions spark prédéfinies

```
display(
  
  apple_store.select(
    
    #Opérations sur des chaines de caractère
    
    F.concat(F.col("prime_genre"), F.lit(" "), F.col("cont_rating")), # Concaténation des champs prime_genre et cont_rating 
    F.split( F.col("track_name"), " " ), # Split du tack_name sur le caractère " "
    F.col("track_name").contains("PAC"), # Test si tack_name contient "PAC" 
    
    #Opérations sur des tableaux
    
    F.size( F.split( F.col("track_name"), " " ) ), # Taille du tableau resultant du split sur le caractère " "
    F.split( F.col("track_name"), " " ).getItem(0), # Récupération du premier élément du tableau 
    F.array_contains( F.split( F.col("track_name"), " " ), "PAC-MAN") #Test si le tableau contient "PAC-MAN" 
    
  )
  
)
```

```
def divide(a, b):
  return a / b

def get_type(elem):
  return str(type(elem))

def split_and_get_first_item(elem, split_char):
  return F.split(elem, split_char).getItem(0)

display(
  
  apple_store.select(
    divide(F.col("rating_count_ver"), F.col("rating_count_tot")),
    F.lit(get_type(F.col("track_name"))),
    split_and_get_first_item(F.col("track_name"), " ")
  )
  
)
```

##C. Les UDF

```
from pyspark.sql.types import * 

def length(string):
  return len(string)

lengthUDF = F.udf(length, IntegerType())

display(
  apple_store.select(
    F.col("track_name"),
    lengthUDF(F.col("track_name"))
  ) 
)
```

```
lengthUDF = F.udf(len, IntegerType())

display(
  apple_store.select(
    F.col("track_name"),
    lengthUDF(F.col("track_name"))
  ) 
)
```

```
def replace(string):
  return string.replace(" ", "_")

replaceUDF = F.udf(replace, StringType())

display(
  apple_store.select(
    F.col("track_name"),
    replaceUDF(F.col("track_name"))
  ) 
)
```

```
def replaceUDF(search, replace):
  def replaceUDFHelper(string):
    return string.replace(search, replace)
  return F.udf(replaceUDFHelper, StringType())

display(
  apple_store.select(
    F.col("track_name"),
    replaceUDF(" ", "^^")(F.col("track_name"))
  ) 
)
```

#4. Aggrégations

```
display(
  apple_store.select(F.count("prime_genre"), F.countDistinct("prime_genre"), F.max("rating_count_tot"))
)
```

```
display(
  apple_store
    .groupBy("prime_genre")
    .agg(
      F.count("track_name"),
      F.countDistinct("cont_rating"),
      F.sum("rating_count_tot"),
      F.mean("price"),
      F.stddev("price")
    )
)
```

```
display(
  apple_store
    .groupBy("prime_genre")
    .agg(F.collect_list("track_name"))
)
```

```
display(
  apple_store.select( 
    F.col("track_name"), 
    F.split( F.col("track_name"), " " ).alias("words"),
    F.explode( F.split( F.col("track_name"), " " ) ).alias("word")
  )
)      
```

```
display(
  
  apple_store.select(
    
    F.col("prime_genre"), 
    F.explode( F.split( F.col("track_name"), " " ) ).alias("word") 
    
  ).groupBy("prime_genre")
   .agg( F.collect_set("word").alias("words") )
  
)      
```

```
display(
  apple_store.groupBy("prime_genre").pivot("cont_rating").count()
)
```

```
display(
  
  apple_store.groupBy("prime_genre")
             .pivot("cont_rating")
             .agg(
               F.countDistinct("track_name"), 
               F.min("size_bytes").alias("min_size"), 
               F.mean("price")
             )
)
```

#5. Window fonctions

```
from pyspark.sql.window import Window

apple_store = apple_store.withColumn(
  "rank",
  F.rank().over( 
    Window.partitionBy("prime_genre")
          .orderBy(F.desc("size_bytes")) 
  )
)

display(apple_store.select("prime_genre", "track_name", "size_bytes", "rank").orderBy("rank"))
```

#6. Jointures et union

```
track_name = apple_store.select("id", "track_name")
prime_genre = apple_store.select("id", "prime_genre")
size = apple_store.select(F.col("id").alias("app_id"), F.col("size_bytes"))
```

```
display(
  track_name.join(prime_genre, "id", "left")
)
```

```
display(
  track_name.join(size, track_name["id"] == size["app_id"], "left")
)
```

```
app_type_navigation = apple_store.filter(F.col("prime_genre") == "Navigation")
app_type_other = apple_store.filter(F.col("prime_genre") != "Navigation")

app_type_navigation.union(app_type_other)
```
