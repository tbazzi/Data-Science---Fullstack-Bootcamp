# 50 Startups

Echauffons nous sur la visualisation avec un dataset simple. Celui-ci décrit le profit de 50 startups en fonction de leurs dépenses dans différents départements. 

**1. Importez le dataset & visualisez le**

```
import numpy as np
import pandas as pd
import seaborn as sns
```

```
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```
dataset = pd.read_csv(fn)
```

```
dataset.head()
```

**2. Avez-vous des valeurs NaN dans le dataset ?**

```
dataset.isnull().any()
```

**3. Faites un nuage de points qui va vous permettre d'exprimer le profit en fonction des dépenses en R&D**

```
sns.relplot(x=dataset.columns[0], y=dataset.columns[-1], data=dataset)
```

**4. Faites de même avec les autres variables quantitatives**

```
sns.relplot(x=dataset.columns[1], y=dataset.columns[-1], data=dataset)
```

```
sns.relplot(x=dataset.columns[2], y=dataset.columns[-1], data=dataset)
```

**5. Regardez maintenant la part de profit en fonction des pays. Vous pourrez l'exprimer par un barplot puis un boxplot. Quel graph est plus pertinent selon vous ?**

```
sns.catplot(x=dataset.columns[3], y=dataset.columns[-1], data=dataset, kind="bar")
```

```
sns.catplot(x=dataset.columns[3], y=dataset.columns[-1], data=dataset, kind="box")
```

**7. Faites une heatmap qui va vous permettre de regarder la corrélation entre les variables. Que pouvez-vous conclure sur les variables les plus influentes sur le profit ?**

```
sns.heatmap(dataset.corr(), annot=True)
```

**8. Est ce que les dépenses en Administration ne seraient pas corrélée avec l'augmentation des dépenses en marketing ou en R&D ? Regardez cela grâce à Seaborn**

```
sns.relplot(x=dataset.columns[0], y=dataset.columns[1], data=dataset, hue=dataset.columns[3])
```
